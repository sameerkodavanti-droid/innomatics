```python
#problem 1
def validate_recharge(amount):
    """
    Validates the recharge amount based on:
    - Minimum amount ₹50
    - Must be one of the valid plans
    """
    valid_plans = [199, 299, 399, 599]

    # Check minimum amount
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Check if plan is valid
    if amount not in valid_plans:
        print("Invalid plan selected.")
        return False

    return True


# Retry logic using while loop
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(recharge_amount):
            print("Recharge successful!")
            break  # Exit loop if valid
        else:
            print("Please try again.\n")

    except ValueError:
        print("Invalid input. Please enter a number.\n")
```

```python
#problem2
def check_inventory(stock_data):
    """
    Checks each product's stock.
    Prints Reorder Alert if stock < 15.
    """

    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Sample inventory dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
#problem3
def process_result(marks_list):
    """
    Calculates average marks.
    Returns Pass if average ≥ 50, else Fail.
    """

    total = 0

    # Calculate total using loop
    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    print(f"Average Marks: {average}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example usage
marks = [60, 45, 70, 50]
result = process_result(marks)
print("Result:", result)
```

```python
#problem4
def calculate_fare(distance, peak_hour):
    """
    Calculates cab fare:
    - Base fare = ₹50
    - ₹12 per km
    - 25% extra during peak hour
    """

    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Apply peak hour surcharge
    if peak_hour:
        fare *= 1.25

    return fare


# Retry option using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

    except ValueError:
        print("Invalid input. Please try again.\n")
```

```python
#problem5
def check_attendance(attendance_list):
    """
    Calculates attendance percentage.
    Returns Eligible if ≥ 75%, else Not Eligible.
    """

    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

```python
#problem6
def check_password_strength(password):
    """
    Checks password strength:
    - Minimum length = 8
    - At least one digit
    - At least one special character (@, #, $)
    """

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"


# Example usage
pwd = input("Enter password: ")
strength = check_password_strength(pwd)
print(strength)
```

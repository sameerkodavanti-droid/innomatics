```python
#problem1: Unique words in a sentence

sentence = "python is easy and python is powerful"
# Split sentence into words
words = sentence.split()
# Convert list to set to get the unique words
uniquewords = set(words)
#now print the results
print("Unique words count:",len(uniquewords))
print("Unique words:", uniquewords)
```

```python
#problem2: Highest salary from employee data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
GotPaidHigh = __builtins__.max(employees, key=employees.get)
salary = employees[GotPaidHigh]
print(f"Highest Salary: {GotPaidHigh} - {salary}")
```

```python
#Problem 3: Find Maximum and Minimum Values

# Given list
values = [45, 22, 89, 10, 66]
# Assume first element as highest and lowest
highest = values[0]
lowest = values[0]
# Loop through the list starting from second element
for num in values[1:]:
    if num > highest:
        highest = num
    if num < lowest:
        lowest = num
print("List:", values)
print("Max:", highest)
print("Min:", lowest)
```

```python
#problem4: count products above a price threshold\

prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000: # Compare each price with 1000
        count += 1
print(f"Products above 1000: {count}")
```

```python
#problem 5 : calculate attendance percentage

attendance = ["P", "P", "A", "P", "P"]
countOfp = attendance.count("P")# Count number of "P" values
total_days = len(attendance)
percentage = (countOfp/total_days)*100 # Calculate percentage
print(f"Attendance Percentage: {percentage}")
```

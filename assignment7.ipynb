## Problem 1: Smart Light Controller

**Problem Statement**
Create a class to control a smart light that can be turned ON or OFF and display its status.

**üåç Real-Time Use**
Smart home automation systems

**üì• Sample Input**
Light Name: Bedroom Light
Action: ON

**üì§ Expected Output**
Bedroom Light is ON

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Sample Usage for Problem 1
bedroom_light = SmartLight("Bedroom Light")
bedroom_light.turn_on()
bedroom_light.display_status()

living_room_light = SmartLight("Living Room Light")
living_room_light.display_status()
living_room_light.turn_on()
living_room_light.display_status()
living_room_light.turn_off()
living_room_light.display_status()
```

## Problem 2: Employee ID Card System

**Problem Statement**
Create a class that stores employee name, ID, and department and displays ID card details.

**üåç Real-Time Use**
Corporate HR systems

**üì• Sample Input**
Employee Name: Rahul
Employee ID: EMP102
Department: IT

**üì§ Expected Output**
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Sample Usage for Problem 2
rahul = Employee("Rahul", "EMP102", "IT")
rahul.display_id_card()

print("\n---")
rohit = Employee("Rohit", "EMP103", "HR")
rohit.display_id_card()
```

## Problem 3: Mobile Contact Record

**Problem Statement**
Create a class to store contact name and phone number and display contact information.

**üåç Real-Time Use**
Mobile contact applications

**üì• Sample Input**
Contact Name: Anita
Phone Number: 9876543210

**üì§ Expected Output**
Contact Saved
Name: Anita
Phone: 9876543210

```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Sample Usage for Problem 3
anita = Contact("Anita", "9876543210")
anita.display_contact()

print("\n---")
david = Contact("David", "1234567890")
david.display_contact()
```

## Problem 4: Product Price Tag Generator

**Problem Statement**
Create a class that stores product name and price and prints a formatted price tag.

**üåç Real-Time Use**
Retail billing systems

**üì• Sample Input**
Product Name: Headphones
Price: 2499

**üì§ Expected Output**
Product: Headphones
Price: ‚Çπ2499

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ‚Çπ{self.price}")

# Sample Usage for Problem 4
headphones = Product("Headphones", 2499)
headphones.print_price_tag()

print("\n---")
keyboard = Product("Mechanical Keyboard", 5999)
keyboard.print_price_tag()
```

## Problem 5: Movie Rating Display System

**Problem Statement**
Create a class that stores movie name and rating and displays movie details.

**üåç Real-Time Use**
Streaming platforms

**üì• Sample Input**
Movie Name: Inception

**üì§ Expected Output**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Sample Usage for Problem 5
inception = Movie("Inception", 4.8)
inception.display_movie_details()

print("\n---")
matrix = Movie("The Matrix", 4.9)
matrix.display_movie_details()
```

## Problem 6: Delivery Address Manager

**Problem Statement**
Create a class that stores customer name and delivery address and prints delivery details.

**üåç Real-Time Use**
Courier and logistics platforms

**üì• Sample Input**
Customer Name: Suman
Address: Hyderabad

**üì§ Expected Output**
Delivery Details
Customer: Suman
Address: Hyderabad

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Sample Usage for Problem 6
suman_delivery = Delivery("Suman", "Hyderabad")
suman_delivery.print_delivery_details()

print("\n---")
rajesh_delivery = Delivery("Rajesh", "Bangalore")
rajesh_delivery.print_delivery_details()
```

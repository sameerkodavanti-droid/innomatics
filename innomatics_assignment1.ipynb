```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
p=0
f=0
for m in marks: #Analyze student results
    if m >= 50:
       p = p+1
    else:
       f =f+1
print("Pass Students:", p)
print("Fail Students:",f)
```

```python
#3. Simple Data Cleaner
raw_names = [" Alice ", "bob", " CHARLIE "]
revised_names = []
for entry in raw_names:
    cleaned = entry.strip().lower() #remove spaces and convert to lowercase
    revised_names.append(cleaned)  #add the cleaned name to the new list
print("Revised names with no extra spaces and lovercase:", revised_names)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg) #calculate the number of characters in the message
    print(f"Message: '{msg}' | Length: {length}")#print message and its length
    if length > 10:
        print("This message is longer than 10 characters")
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
ec = logs.count("ERROR")#counting the error entries
print(f"Total Error Count: {ec}")
```

Problem 1: Social Media – Post Engagement Analyzer


```python
def check_post_virality(engagement_metrics):
    # Summing up all likes from the list of interactions
    accumulated_likes = sum(engagement_metrics)

    # Determining if the post reached viral status
    if accumulated_likes >= 1000:
        label = "Viral Post"
    else:
        label = "Normal Engagement"

    return accumulated_likes, label

# Sample Data
likes_data = [400, 300, 350]
total, status = check_post_virality(likes_data)

print(f"Total Likes: {total}")
print(f"Post Status: {status}")
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def evaluate_precipitation(monthly_records):
    # Calculating the mean rainfall over the period
    avg_rain = sum(monthly_records) / len(monthly_records)

    # Threshold check for agricultural adequacy
    if avg_rain >= 70: # Assuming 70 as a general required level based on context
        result = "Adequate Rainfall"
    else:
        result = "Inadequate Rainfall"

    return avg_rain, result

# Sample records
rainfall_stats = [80, 65, 71]
average, adequacy = evaluate_precipitation(rainfall_stats)

print(f"Average Rainfall: {average:.1f}")
print(f"Rainfall Status: {adequacy}")
```

Problem 4: Social Media – Duplicate Account Detection

```python
def find_cloned_profiles(user_list):
    # Converting list to a set to find unique entries
    unique_profiles = set(user_list)

    # Comparing lengths to identify if duplicates were filtered out
    if len(unique_profiles) != len(user_list):
        return "Yes"
    return "No"

# Testing the detection
profile_names = ["user1", "dev_coder", "user1", "alpha_tester"]
has_duplicates = find_cloned_profiles(profile_names)

print(f"Duplicate Accounts Found: {has_duplicates}")
```

Problem 5: Healthcare – Appointment Eligibility Checker

Problem 6: Agriculture – Premium Crop Price Filter

```python
def extract_high_value_crops(price_catalog):
    # Using list comprehension to filter prices above 2000
    top_tier_prices = [p for p in price_catalog if p > 2000]
    return top_tier_prices

# Catalog prices
market_prices = [1500, 2500, 1800, 3200, 900]
premium_list = extract_high_value_crops(market_prices)

print(f"Premium Crops: {premium_list}")
```

Problem 7: System Monitoring – Application Health Checker


```python
def assess_server_health(detected_errors):
    # Categorizing health based on error frequency
    if detected_errors == 0:
        health_state = "Healthy"
    elif detected_errors <= 5:
        health_state = "Minor Issues"
    else:
        health_state = "Critical Issues"
    return health_state

# Diagnostic run
issue_count = 7
system_report = assess_server_health(issue_count)

print(f"Error Count: {issue_count}")
print(f"System Status: {system_report}")
```

Problem 8: Banking – Daily Transaction Limit Checker


```python
def validate_banking_limit(requested_sum):
    CEILING_LIMIT = 50000

    # Verify if the request is within the daily cap
    if requested_sum <= CEILING_LIMIT:
        decision = "Approved"
    else:
        decision = "Rejected"
    return decision

# Transaction attempt
withdrawal_amount = 60000
outcome = validate_banking_limit(withdrawal_amount)

print(f"Transaction Amount: {withdrawal_amount}")
print(f"Transaction Status: {outcome}")
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_qualification(session_logs):
    # Calculating the ratio of attended sessions
    total_classes = len(session_logs)
    attended_count = session_logs.count(1) # Assuming 1 is present, 0 is absent
    ratio = (attended_count / total_classes) * 100

    # Academic eligibility rule
    if ratio >= 75:
        permit = "Eligible"
    else:
        permit = "Not Eligible"

    return ratio, permit

# Student record (1: Present, 0: Absent)
attendance_data = [1, 1, 0, 1, 1]
percent, status = check_exam_qualification(attendance_data)

print(f"Attendance Percentage: {percent}%")
print(f"Exam Eligibility: {status}")
```

Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_power_invoice(units_used):
    cost = 0

    # Applying tiered billing rates
    if units_used <= 100:
        cost = units_used * 3
    elif units_used <= 200:
        cost = (100 * 3) + ((units_used - 100) * 5)
    else:
        cost = (100 * 3) + (100 * 5) + ((units_used - 200) * 7)

    # Categorizing based on final invoice amount
    if cost < 500:
        bracket = "Low Usage"
    elif cost <= 1500:
        bracket = "Moderate Usage"
    else:
        bracket = "High Usage"

    return cost, bracket

# Consumption check
meter_reading = 250
total_bill, usage_tier = calculate_power_invoice(meter_reading)

print(f"Units Consumed: {meter_reading}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {usage_tier}")
```
